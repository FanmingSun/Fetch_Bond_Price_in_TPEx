```python
from urllib.error import HTTPError
def readUrl(date):
  excel_url = f"https://www.tpex.org.tw/storage/bond_zone/tradeinfo/govbond/{date[:4]}/{date[:6]}/BDdys01a.{date}-C.xls"
  try :
    df = pd.read_excel(excel_url)
    return df
  except pd.errors.ParserError as e:
    return None
  except HTTPError as e:
    return None

result_df = readUrl('20231024')

if result_df is not None:
    # Process or use the DataFrame as needed
    result_df.dropna(inplace=True)
    print(result_df.head())
```

```python
#設定資料下載期間
date = ['20231024','20231023','20231022','20231020','20231019','20231018']
```

```python
combined_df = pd.DataFrame()
for d in date:
  result_df = readUrl(d)
  if result_df is not None:
    result_df.dropna(inplace=True)
    result_df = result_df.iloc[1:, lambda df: [0 ,5]]
    result_df["Date"] = d
    result_df = result_df.rename(columns = {"等殖成交行情表(買賣斷)\nDaily Prices & Volume－Outright Purchase & Sales (Electronic Bonds Trading System)": "Bond",'Unnamed: 5':"Price" })
    result_df.set_index("Date")
    combined_df = pd.concat([combined_df, result_df])
```
